```python
import pandas as pd
import os
import numpy as np

%load_ext autoreload
%autoreload 2
```

```python
data_path = 'data'
input_csv_name = 'val.csv' 
ifile_name = os.path.join(data_path + os.sep + input_csv_name) #Raw SO scrapped data
```

```python
df = pd.read_csv(ifile_name)
df
```

### BodyLength
For each answer, we get size of answer body in words

```python
length_list = []
from body_util import clear_tags
datapoints = df.shape[0]
for i in range(datapoints): 
    answer_text = clear_tags(df['Body'][i])
    length_list.append(len(answer_text.split()))

df['BodyLength'] = length_list
df
```

### Code
For each answer, we calculate number of code blocks

```python
codes_inline = []
codes_pre_line = []
codes_pre_count = []
datapoints = df.shape[0]
for i in range(datapoints): 
    answer_text = df['Body'][i]
    code_start = 0
    code_end = 0
    code_pre_line = 0 #Number of code blocks
    code_pre_count = 0 #Number of lines in code blocks
    code_inline = 0 #Number of inline code snippets
    while True:
        code_start = answer_text.find('<code>', code_end)
        if code_start == -1:
            break
        code_end = answer_text.find('</code>', code_start)
        code = answer_text[code_start: code_end+7]
        # Here, code is a string which is of the format <code>.....</code>
        if answer_text[code_start-5: code_start] == '<pre>':
            code_line = code.split('\n')
            code_line_count = len(code_line)-1 # Last line as </code>
            code_pre_line += code_line_count
            code_pre_count += 1
        else:
            code_inline += 1
    codes_inline.append(code_inline)
    codes_pre_line.append(code_pre_line)
    codes_pre_count.append(code_pre_count)

df['InlineCode'] = codes_inline
df['BlockCode'] = codes_pre_count
df['BlockCodeLine'] = codes_pre_line
df
```

### Hyperlink
For each post, we calculate the number of links referenced

```python
hyperlinks = []
datapoints = df.shape[0]
for i in range(datapoints):
    body = df['Body'][i]
    hyperlink_pos = -1
    hyperlink = 0
    while True:
        hyperlink_pos = body.find('<a href=', hyperlink_pos+1)
        if hyperlink_pos == -1:
            break
        hyperlink += 1
        # print(i, body[hyperlink_pos:hyperlink_pos + 50])
    hyperlinks.append(hyperlink)

df['Hyperlink'] = hyperlinks
df
```

### Edit
For each post, we check if it was edited

```python
edits = []
datapoints = df.shape[0]
for i in range(datapoints):
    last_edit_date = df['LastEditDate'][i] # str or float (np.nan)
    if isinstance(last_edit_date, float):
        if np.isnan(last_edit_date):
            edits.append('0')
        else:
            edits.append('1')
    else:
        edits.append('1')
df['Edit'] = edits
df
```

### Label
For each post, we classify if it is the correct answer

```python
labels = []
datapoints = df.shape[0]
for i in range(datapoints):
        
    answer_id = df['Id'][i] # int
    accepted_id = df['QAcceptedAnswerId'][i] # float (np.nan)
    if np.isnan(accepted_id):
        labels.append(np.nan)
    else:
        if answer_id == int(accepted_id):
            labels.append('1')
        else:
            labels.append('0')
df['Label'] = labels
df
```

### Unigram + Bigram scores
First, we generate a corpus of well written answers

```python
from ngram import gen_corpus
from os.path import exists

corpus_file = 'data/corpus.txt' #This is the output which will be generated
if input_csv_name == 'train.csv':
    gen_corpus(df[df.Label=='1'].Body.tolist(),corpus_file)
else:
    print ("NGram corpus only for train data")
    if not exists(corpus_file):
        raise Exception("No corpus present. Generate corpus with training data first")
```

Now, we use the corpus file to generate a language model. Then we can calculate unigram and bigram costs

```python
from ngram import make_language_model
from ngram import SENTENCE_BEGIN

unigram_cost, bigram_cost = make_language_model(corpus_file)

ug = []
bg = []

datapoints = df.shape[0]

for i in range(datapoints):
        line = df['Body'][i]
        # Unigram cost.
        grams=tuple(line)
        cost=0
        for j in range(len(grams)):
            cost+=unigram_cost(grams[j])
        cost=cost/len(grams)
        ug.append(cost)

        cost=bigram_cost(SENTENCE_BEGIN, grams[0])
        for j in range(1,len(grams)):
            cost+=bigram_cost(grams[j-1], grams[j])
        cost=cost/len(grams)
        bg.append(cost)


    # create a output panda dataframe
df['UnigramCost'] = ug
df['BigramCost'] = bg
df
```

### Output data

```python
output = df[['Label','Edit','CommentCount','BodyLength', 'Reputation','Views','UpVotes','DownVotes','InlineCode','BlockCode','BlockCodeLine','Hyperlink','UnigramCost','BigramCost']]
output
```

```python
output_csv_name = input_csv_name.replace('.csv','_feature.csv')
ofile_name = os.path.join(data_path + os.sep + output_csv_name) #Feature extracted SO data
output.to_csv(ofile_name)
```
