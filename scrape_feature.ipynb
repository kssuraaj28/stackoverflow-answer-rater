```python
import pandas as pd
import os
import numpy as np
```

```python
raw_data_path = 'SO_raw_data'
csv_name = 'Feb-2020-day.csv' 
ifile_name = os.path.join(raw_data_path + os.sep + csv_name) #Raw SO scrapped data
```

```python
df = pd.read_csv(ifile_name)
df
```

### BodyLength
For each answer, we get size of answer body in words

```python
length_list = []
datapoints = df.shape[0]
for i in range(datapoints): 
    answer_text = df['Body'][i]
    length_list.append(len(answer_text.split()))

df['BodyLength'] = length_list
df
```

### Code
For each answer, we calculate number of code blocks

```python
codes_inline = []
codes_pre_line = []
codes_pre_count = []
datapoints = df.shape[0]
for i in range(datapoints): 
    answer_text = df['Body'][i]
    code_start = 0
    code_end = 0
    code_pre_line = 0 #Number of code blocks
    code_pre_count = 0 #Number of lines in code blocks
    code_inline = 0 #Number of inline code snippets
    while True:
        code_start = answer_text.find('<code>', code_end)
        if code_start == -1:
            break
        code_end = answer_text.find('</code>', code_start)
        code = answer_text[code_start: code_end+7]
        # Here, code is a string which is of the format <code>.....</code>
        if answer_text[code_start-5: code_start] == '<pre>':
            code_line = code.split('\n')
            code_line_count = len(code_line)-1 # Last line as </code>
            code_pre_line += code_line_count
            code_pre_count += 1
        else:
            code_inline += 1
    codes_inline.append(code_inline)
    codes_pre_line.append(code_pre_line)
    codes_pre_count.append(code_pre_count)

df['InlineCode'] = codes_inline
df['BlockCode'] = codes_pre_count
df['BlockCodeLine'] = codes_pre_line
df
```

### Hyperlink
For each post, we calculate the number of links referenced

```python
hyperlinks = []
datapoints = df.shape[0]
for i in range(datapoints):
    body = df['Body'][i]
    hyperlink_pos = -1
    hyperlink = 0
    while True:
        hyperlink_pos = body.find('<a href=', hyperlink_pos+1)
        if hyperlink_pos == -1:
            break
        hyperlink += 1
        # print(i, body[hyperlink_pos:hyperlink_pos + 50])
    hyperlinks.append(hyperlink)

df['Hyperlink'] = hyperlinks
df
```

### Edit
For each post, we check if it was edited

```python
edits = []
datapoints = df.shape[0]
for i in range(datapoints):
    last_edit_date = df['LastEditDate'][i] # str or float (np.nan)
    if isinstance(last_edit_date, float):
        if np.isnan(last_edit_date):
            edits.append('0')
        else:
            edits.append('1')
    else:
        edits.append('1')
df['Edit'] = edits
df
```

### Label
For each post, we classify if it is the correct answer

```python
edits = []
labels = []
datapoints = df.shape[0]
for i in range(datapoints):
        
    answer_id = df['Id'][i] # int
    accepted_id = df['QAcceptedAnswerId'][i] # float (np.nan)
    if np.isnan(accepted_id):
        labels.append(np.nan)
    else:
        if answer_id == int(accepted_id):
            labels.append('1')
        else:
            labels.append('0')
df['Label'] = labels
df
```

### Output data

```python
output = df[['Label','Edit','CommentCount','BodyLength', 'Reputation','Views','UpVotes','DownVotes','InlineCode','BlockCode','BlockCodeLine','Hyperlink']]
output
```

```python
parsed_data_path = 'data'
csv_name = csv_name.replace('.csv','_feature.csv')
ofile_name = os.path.join(parsed_data_path + os.sep + csv_name) #Feature extracted SO data
output.to_csv(ofile_name)
```
